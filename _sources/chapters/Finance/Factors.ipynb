# What Are Factors?

Factors are characteristics or signals that help explain asset returns. They serve both as **risk indicators** and as **return drivers**. Academics emphasize their ability to explain **expected returns** and remove known sources of variation so that only ‚Äúnew‚Äù sources of alpha remain. Practitioners value them for improving **performance measurement**, **risk management**, and **portfolio construction**.

We can group factors based on what they aim to capture:

## üåç Market & Sector Membership
- **Market**: Whether the asset belongs to a region (e.g., US, China). Encoded as 1 if present, 0 otherwise.
- **Sector**: Broad industry grouping (e.g., tech, healthcare). Can be split into 5 to 100+ categories.
- Works beyond equities‚Äîapplicable to credit markets, bonds, etc.

These ‚Äúmembership‚Äù factors help isolate common shocks shared within countries or sectors.

## üßæ Accounting-Based Factors
These aim to proxy for fundamentals and predict returns:

- **Value (Book-to-Market)**: Cheap stocks (high book relative to price) have historically earned higher returns. This remains one of the most studied and used factors in practice.
- **Profitability / Quality**: Gross profitability (Revenue ‚Äì COGS) / Assets helps separate high-quality stocks from deteriorating ones. Other signals include ROE, margins, earnings stability, accruals, and low debt levels.
- **Investment**: Firms that invest less (e.g., low asset growth, low capital expenditures) tend to outperform, potentially because high investment reflects low marginal value or managerial empire-building.
- **Composite Quality Scores**: Many practitioners combine profitability, stability, leverage, and growth into a single ‚Äúquality‚Äù score for screening or weighting.

Combining quality with value is especially powerful: among firms with similar value metrics, those with high profitability tend to outperform.

## üìâ Return-Based Factors
These come from past return patterns and risk exposures:

- **Beta**: Measures sensitivity to the market factor. The CAPM suggests a linear relation between beta and expected return, but empirically this relationship often breaks down.
- **Volatility**: High total or idiosyncratic volatility is often associated with *lower* expected returns. Low-volatility stocks have outperformed in many global markets.
- **Short-Term Reversal**: Stocks with very poor short-term performance tend to revert upwards, likely due to liquidity shocks, overreactions, or temporary flows.
- **Momentum**: One of the strongest factors. Past 12-month return (excluding the most recent month) predicts future returns. Often interpreted as a proxy for underreaction or investor herding.
- **Skewness / Tail Risks**: Assets with negatively skewed returns (lottery-like) may be overpriced due to behavioral demand.

## üîÅ Trading-Based Factors
These reflect investor behavior and flow-driven dynamics:

- **Short Interest**: High short interest or high borrow cost signals pessimism and often predicts underperformance.
- **Active Ownership**: Stocks targeted by activists or insiders tend to outperform, especially if tied to governance or restructuring.
- **Liquidity / Amihud Illiquidity**: Illiquid stocks earn higher returns, possibly compensating for trading costs and risk.
- **Turnover / Share Issuance**: High share issuance and high turnover have been linked to lower future returns, possibly reflecting overpricing.

Crowdedness matters: signals can lose predictive power when overused. Many trading-based factors capture frictions or flow-induced mispricing that decay with scale.


## Important Note
Even if a factor does not reliably predict *expected returns*, it may still explain *co-movement* across assets. That alone makes it essential for risk management and performance attribution. A robust understanding of both risk-based and behavioral interpretations helps in building resilient portfolios and separating true alpha from systematic exposure.

## üíº Value Investing ‚Äì Deep Dive

Value investing is one of the most widely studied and implemented quantitative strategies.

### üîé What is it?
A strategy that tilts the portfolio toward stocks that appear *cheap* relative to a fundamental anchor‚Äîand away from those that appear *expensive*.

Mathematically:
$$
W_t \propto \frac{F_t}{P_t}
$$
Where \( F_t \) is a proxy for a firm‚Äôs fundamental value and \( P_t \) is its price.

Common choices for \( F_t \):
- Book Value
- Earnings
- Dividends
- Cash Flows
- Sales

This is the cross-sectional counterpart to time-series strategies that time the market based on aggregate valuation metrics like the dividend yield. However, cross-sectional value strategies tend to be more robust‚Äîeven if they can underperform for long periods (as seen in the 2010s).










### üìö Academic Foundations
- **[Lakonishok, Shleifer, and Vishny (1994) ](https://www8.gsb.columbia.edu/sites/valueinvesting/files/files/CIER.pdf)**: Behavioral view‚Äîinvestors overreact to past growth or bad news. Value stocks exploit these biases.
- **[Fama and French (1996)](https://onlinelibrary.wiley.com/doi/pdfdirect/10.1111/j.1540-6261.1996.tb05202.x)**: Rational risk-based view‚Äîvalue stocks are riskier, consistent with multifactor models.

### üß† Why Might It Work?

**1. Mispricing (Behavioral Story)**
- Investors extrapolate trends or overreact to news.
- Growth stocks become overpriced (‚Äúglamorous‚Äù); value stocks are neglected (‚Äúugly‚Äù).
- Contrarian rebalancing can capture mean-reversion in sentiment.

**2. Risk (Rational Story)**
- Value firms may be fundamentally riskier: more distress-prone, less flexible, or more sensitive to macro shocks.
- This explains why value did poorly in crises like 2008 and 2020.

**Empirical Challenge to CAPM:**
- In CAPM, higher expected return implies higher beta.
- But as firms transition between value/growth portfolios, their beta doesn‚Äôt change accordingly.
- This inconsistency suggests that beta alone can‚Äôt explain value spreads.



### ‚öñÔ∏è Summary
Most researchers believe there‚Äôs a blend of both:
- **Behavioral** elements (investor overreaction, underreaction, salience).
- **Risk-based** explanations (distress risk, macro exposure).

Despite recent underperformance, value remains foundational in both academic research and institutional practice.

## üìà Return-Based Strategies

**Momentum** is the tendency for stocks that have performed **relatively well** in the recent past to continue performing well in the near future‚Äîand likewise for recent underperformers. This is a **cross-sectional** phenomenon: it‚Äôs about relative strength across stocks, not absolute performance.

A classic **momentum strategy** buys recent winners and shorts recent losers. Over the 1963‚Äì1990 period, Jegadeesh and Titman (1993) showed that a strategy ranking stocks by past 6-to-12-month returns and taking long/short positions in the top/bottom deciles delivered **abnormal returns of 12% per year**.

### ‚öñÔ∏è Momentum vs. Value

While **value** strategies buy beaten-down stocks based on fundamentals, **momentum** buys recent outperformers based on price trends. In fact, value and momentum often load on opposite ends of the performance spectrum:

- **Value stocks** have low recent returns (hence low prices vs. fundamentals)
- **Momentum stocks** have high recent returns (regardless of valuation)


### ü§î Why Might It Work?

**Behavioral Explanation**:
- Investors are slow to fully process new information
- News and sentiment get incorporated into prices gradually
- Examples include political developments, earnings trends, or macro regime shifts (e.g., COVID, elections)

**Rational Explanation**:
- Stocks with similar momentum may share exposure to unobserved risk factors
- Momentum return could reflect compensation for time-varying risk exposures
- However, it's hard to find a well-defined risk factor that explains the magnitude and persistence of the momentum premium

Most academics lean toward behavioral explanations, though some admit rational risks may help explain co-movement.

‚ö†Ô∏è **Caveat**: Momentum **crashes hard in high-volatility environments**‚Äînotably in 2008 and 2020. So while profitable, it is **not low risk**.

| Horizon | Pattern | Key Papers |
|--------|---------|------------|
| **Long-term (3‚Äì5 years)** | Reversal (contrarian effect) | DeBondt & Thaler (1985) |

Returns tend to reverse over very short and very long horizons‚Äîbut continue in the intermediate range. This supports the idea of **temporary mispricing** that corrects over time.

## üõ°Ô∏è Defensive Investing

**Core idea**:  
Don‚Äôt chase high-return signals‚Äîtarget **low-risk** assets instead. Defensive investing is about recognizing that **risk is more predictable than return** and using that predictability to build superior portfolios.

> "Why look for high-return assets if lower-risk ones consistently outperform on a risk-adjusted basis?"


### ü§î Motivation

- **Traditional theory** (e.g., CAPM) links higher expected return to higher beta.
- But in the data, **beta is not reliably compensated**.
- Instead, **low-risk stocks often outperform high-risk ones** on a risk-adjusted basis.
- The **tangency portfolio** (true mean-variance optimal portfolio) is unobservable‚Äîreal-world risk-return trade-offs may deviate from theory.


### üìâ Three Key Low-Risk Anomalies

#### **Betting Against Beta (BAB)**  
**Source**: Frazzini & Pedersen (2014)

- Strategy: Long **low-beta** stocks, short **high-beta** stocks, leverage up to equalize risk.
- Empirical finding: Low-beta stocks earn **high alphas**, high-beta stocks earn **very negative alphas**.

üìä BAB Sharpe Ratio Evidence:

> ‚úÖ This contradicts CAPM predictions and reveals an exploitable anomaly.



---

#### **Low Volatility Anomaly**  
**Source**: Ang, Hodrick, Xing, Zhang (2006, 2009)

- Stocks with **lower total or idiosyncratic volatility** have historically outperformed riskier stocks.
- Intuition: Investors overpay for volatility (e.g., ‚Äúlottery tickets‚Äù).

---

####  **Hedging Risk Factors**  
**Source**: Hersovick, Moreira & Muir (2018)

Rather than targeting individual stocks, this strategy builds **minimum-variance portfolios** relative to a factor model.

---

### üßÆ How It Works (HMM Method)

1. **Choose a factor model**:  
   Let $F_t $ be a vector of factor returns (e.g., market, value, momentum, etc.).

2. **Run time-series regressions** for each stock $i $:  
  $$
   R^e_{i,t} = \alpha_i + \beta_{i,F} F_t + \varepsilon_{i,t}
   $$

3. **Build the total return covariance matrix**:  
   - Stack all $\beta_{i,F} $ into a matrix  
   - Assume residuals $\varepsilon_{i,t} $ are uncorrelated across stocks (diagonal $\text{Var}(\varepsilon) $)  
   - Use:
    $$
     \text{Var}(R^e) = \beta \cdot \text{Var}(F_t) \cdot \beta^\top + \text{Var}(\varepsilon)
     $$

4. **Construct the minimum variance portfolio**:
  $$
   w \propto \text{Var}(R^e)^{-1} \cdot \mathbf{1}
   $$

This yields a **zero-cost, risk-minimized portfolio** that aggressively hedges exposure to systematic risk.

- This creates **overpricing** in high-beta stocks and **underpricing** in low-beta stocks.
- Defensive strategies exploit that mispricing, effectively **betting against risk**.

## üèÜ Quality Investing

**What is ‚Äúquality‚Äù investing?**  
It‚Äôs investing in businesses you‚Äôre willing to pay more for‚Äîall else equal. These firms are often fundamentally stronger: profitable, well-managed, and resilient.

### üîç How Is Quality Defined?

There‚Äôs no single definition, but common indicators include:

- **Profitability** (gross profits, margins, ROE)
- **Safety** (low leverage, stable earnings)
- **Good governance** (shareholder-aligned incentives)
- **Growth** (asset growth, earnings growth)
- **High payout ratios**
- **Creditworthiness**
- **Well-managed operations**
- **Accounting quality** (low accruals, consistent earnings)

> In practice, many investors use **composite quality scores** that combine multiple of the above.


### üîÅ Quality vs Value

- Value = **Buy what's cheap**  
- Quality = **Buy what‚Äôs strong**

Yet the two are often linked:

> ‚ÄúBuying high-quality firms without paying a premium is another form of value investing.‚Äù

üí¨ Warren Buffett:  
> ‚ÄúIt‚Äôs far better to buy a wonderful business at a fair price than a fair business at a wonderful price.‚Äù


### üí° Simple Quality Signal: Gross Profitability

One of the most effective and intuitive quality signals is:

$$
\text{Gross Profitability} = \frac{\text{Revenue} - \text{COGS}}{\text{Assets}}
$$

‚úÖ Introduced in [Novy-Marx (2013)](http://rnm.simon.rochester.edu/research/OSoV.pdf), this measure predicts returns even when controlling for book-to-market.

### üìâ Why Might Quality Work?

1. **Rational view**:  
   - Productive firms (profitable) should command higher expected returns.
   - Investors might demand higher returns for riskier, lower-quality companies.

2. **Behavioral view**:  
   - Investors overvalue ‚Äúglamour‚Äù growth firms and neglect boring but profitable ones.
   - Mispricing arises when investors chase growth narratives without accounting for real productivity.



### üß™ Empirical Evidence

- Gross profitability strongly predicts returns‚Äîeven **after** controlling for book-to-market.
- Profitable firms:
  - Have higher average returns
  - Tend to have **lower B/M ratios**
  - Are larger (higher market caps)

üìä Evidence from Novy-Marx (2013):  
![Profitability Alpha](https://github.com/amoreira2/Fin418/blob/main/assets/plots/othertradingstrategiesB2.jpg?raw=1)

- The **Profitable Minus Unprofitable (PMU)** portfolio earns a **high alpha** even relative to the Fama-French 3-factor model.
- Fama and French later incorporated this into their 5-factor model as **RMW (Robust Minus Weak)**.
- The **alpha is large** in part because quality is negatively correlated with value‚Äîquality strategies tend to short cheap but low-quality stocks.



---

### üß© Strategic Role of Quality

- **Quality is a valuable hedge for value**: it outperforms precisely when value underperforms.
- Value strategies tend to hold low-quality names. Quality strategies often avoid them.
- Combining value and quality enhances Sharpe and reduces downside risk.

üìä This synergy helps explain several other anomalies:
- Return on Equity (ROE)
- Net stock issuance
- Default risk
- Organizational capital
- Earnings stability
- Market power

## üìù Key Takeaways

Based on the discussion in the notebook, here are the key takeaways:

*   **Factors as Risk and Return Drivers**: Factors are characteristics that explain asset returns, serving as both risk indicators and drivers of expected returns.
*   **Factor Groupings**: Factors can be grouped into categories like Market & Sector Membership, Accounting-Based, Return-Based, and Trading-Based, each capturing different aspects of asset behavior.
*   **Value Investing**: This strategy focuses on buying stocks cheap relative to their fundamentals. It is supported by both behavioral (investor overreaction) and rational (risk-based) explanations, although it can underperform for extended periods.
*   **Momentum Investing**: This strategy involves buying past winners and shorting past losers based on recent price trends. It is primarily explained by behavioral factors (slow information processing) but carries significant risk, especially during high-volatility periods.
*   **Defensive Investing**: This approach targets low-risk assets, challenging the traditional view that higher returns require higher beta. Anomalies like Betting Against Beta (BAB) and Low Volatility demonstrate that low-risk stocks can achieve high risk-adjusted returns, often due to leverage constraints and investor preferences for high-beta assets.
*   **Quality Investing**: This strategy focuses on investing in fundamentally strong businesses with characteristics like high profitability, safety, and good governance. Quality is often linked to value and can act as a hedge against value underperformance. Simple metrics like Gross Profitability are effective quality signals.
*   **Factor Synergy**: Combining different factors, such as value and quality, can enhance portfolio performance and risk management. Some factors, like profitability, can explain the alphas of other seemingly unrelated strategies.
*   **Importance of Risk Management**: Even if a factor doesn't reliably predict expected returns, it can still explain co-movement across assets, making it crucial for risk management and performance attribution.
